### Tasks

2. Print the orthographic representation, lemma, pos, entity type
3. Print all the entities
4. Print all the sentences

6. Sort the dictionary in descending order

8. Draw a bar plot with top 20 words based on frequency


#### Step 0

```python
text_data = open('opioid_epidemic.txt', encoding='utf-8').read()
```

```python
print(text_data) 
# Without print the \n character will not be implemented
```

#### Step 1 - Creating Tokens

```python
parsed_text = nlp(text_data)
print(type(parsed_text))
```

#### Step 2 - Printing the orthographic representation, lemma, pos, entity type

```python
for token in parsed_text:
    print(token.orth_, token.lemma_, token.pos_, token.ent_type_)
    
```

#### Step 3 - Printing all the entities

```python
for ent in parsed_text.ents:
    print(ent,' ---    ', ent.label_)
```

#### Step 4 - Printing all sentences

```python
for idx,sen in enumerate(parsed_text.sents, start = 1):
    print(idx,':', sen)
```

#### Step 5 - Dictionary with words and frequency

##### Method 1

```python
dict_words = {}
for token in parsed_text:
    lemma = (token.lemma_).lower()
    if lemma in dict_words.keys():
        dict_words[lemma]+=1
    else:
        dict_words[lemma] = 1
```

```python
# print(dict_words)
dict_words.items()
```

##### Method 2  - using  Counter object 

```python
from collections import Counter
```

```python
dict_words_2 = Counter()

for token in parsed_text:
    lemma = (token.lemma_).lower()
    dict_words_2[lemma]+=1
    
```

```python
dict_words_2.items() 
```

##### The structure of both dictionary item is same - list of tuples 

#### Step 6 - Sort the dictionary in descending order

```python
from operator import itemgetter
# sorted_list_words = sorted(dict_words_2.items(),key = itemgetter(1), reverse=True)
sorted_list_words = sorted(dict_words_2.items(),key = lambda x : x[1], reverse=True)
      
```

```python
sorted_list_words
```

#### Step 7 - Creating a new dictionary based on tokens which are not stop-word, punctuation, spaces 

```python
dict_key_words = Counter()
for token in parsed_text:
    lemma = (token.lemma_).lower()
    if not (nlp.vocab[lemma].is_stop or token.pos_ == 'PUNCT' or token.pos_ == 'SPACE'):
        dict_key_words[lemma] +=1
```

```python
dict_key_words
```

#### Step 8 - Bar plot with top 20 words

```python
list_key_words_sorted = sorted(dict_key_words.items(),key = lambda x: x[1], reverse= True)
```

```python
top_count = 20
list_top_words = [word for word, count in list_key_words_sorted[:20]]
list_top_count = [count for word, count in list_key_words_sorted[:20]]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
plt.style.use('ggplot')
plt.bar(list_top_words, list_top_count)
plt.xlabel("Words")
plt.ylabel("Counts")
plt.xticks( rotation = 90)
```
